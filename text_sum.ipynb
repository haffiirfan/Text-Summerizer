```python
from transformers import pipeline
summarizer=pipeline('summarization',model='t5-small',tokenizer='t5-small')
```

```python
def summarize_text(text,max_input_length=512,max_output_length=50):
  input_text='summarize: '+text.strip()
  summary=summarizer(input_text,max_length=max_output_length,min_length=10,do_sample=False,truncation=True)
  return summary[0]['summary_text']
```

```python
paragraph='''Cricket is one of the most popular games not just in India but across the world.
 It is played between two teams consisting of a total of eleven players each. Watching others
 play cricket is as exciting as playing it. You will find below a number of short and long
 paragraphs on Cricket. We hope these Cricket paragraphs will help students in completing their
 school assignments. These will also help children to write and read out paragraphs with simple
 words and small sentences. Students can select any paragraph on Cricket according to their
 particular requirement.'''
print("Original Paragraph:\n", paragraph)
print("\nSummary:\n", summarize_text(paragraph))
```
